```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the coffee dataset
coffee_data = pd.read_csv('/content/Project[1].csv')

# Data Cleaning and Wrangling
# Check for missing values
print("Missing Values:")
print(coffee_data.isnull().sum())

# Perform data cleaning and wrangling as needed
# Example: Handle missing values, data transformation, reshape, etc.

# Descriptive Statistics
# Calculate descriptive statistics
print("Descriptive Statistics:")
print(coffee_data.describe())

# Data Visualization
# Example visualizations

# Histograms
plt.figure(figsize=(10, 6))
sns.histplot(coffee_data['Total_Bill'], bins=20, kde=True)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()

# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Day Name', y='Total_Bill', data=coffee_data)
plt.title('Total Bill on Different Days')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()

# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Hour', y='Total_Bill', data=coffee_data)
plt.title('Total Bill vs. Hour')
plt.xlabel('Hour')
plt.ylabel('Total Bill')
plt.show()

# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(coffee_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Categorical Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='product_category', y='Total_Bill', data=coffee_data)
plt.title('Total Bill by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Bill')
plt.xticks(rotation=45)
plt.show()

# Insightful Analysis
# Example insights

# Insight 1: Analyze the relationship between product category and total bill
avg_bill_per_category = coffee_data.groupby('product_category')['Total_Bill'].mean()
print("Average Total Bill per Product Category:")
print(avg_bill_per_category)

# Insight 2: Analyze sales variation across different days of the week
avg_sales_per_day = coffee_data.groupby('Day Name')['Total_Bill'].mean()
print("Average Total Bill per Day:")
print(avg_sales_per_day)

# Conditions
# Explore specific conditions or hypotheses

# Condition 1: Identify the top-selling products or categories
top_selling_products = coffee_data.groupby('product_type')['Total_Bill'].sum().nlargest(3)
print("Top Selling Product Types:")
print(top_selling_products)

# Condition 2: Analyze the impact of transaction quantity on total bill
plt.figure(figsize=(10, 6))
sns.scatterplot(x='transaction_qty', y='Total_Bill', data=coffee_data)
plt.title('Total Bill vs. Transaction Quantity')
plt.xlabel('Transaction Quantity')
plt.ylabel('Total Bill')
plt.show()


```
